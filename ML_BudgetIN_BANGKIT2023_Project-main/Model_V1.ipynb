```python
price_ranges = {
    'Sangat Murah': (0, 15000),
    'Murah': (15000, 40000),
    'Mahal': (40000, 100000),
    'Sangat Mahal': (100000, np.inf)
}

price_ranges.items()
```

```python
def get_price_range(row):
    for category, (min_val, max_val) in price_ranges.items():
        if min_val <= row['min'] <= max_val and min_val <= row['max'] <= max_val:
            return category
    return 'Lainnya'

aggregated['kategori_harga'] = aggregated.apply(get_price_range, axis=1)
aggregated
```

Main ( For now using SIMPLE MODEL )

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Flatten, Dense, Concatenate
from tensorflow.keras.losses import MeanSquaredError

# Example dataset of eating places
places = [
    {'name': 'Restaurant A', 'max': 30000.0, 'merchant_area': 'jakarta', 'category_price': 'Murah'},
    {'name': 'Restaurant B', 'max': 50000.0, 'merchant_area': 'medan', 'category_price': 'Mahal'},
    {'name': 'Restaurant C', 'max': 25000.0, 'merchant_area': 'surabaya', 'category_price': 'Sangat Murah'},
    {'name': 'Restaurant D', 'max': 80000.0, 'merchant_area': 'surabaya', 'category_price': 'Sangat Mahal'},
    # Add more eating places and their attributes
]

# Example user budget and preferences
user_budget = 30000.0
merchant_area_def = 'surabaya'

# Preprocess the data and extract relevant features
merchant_mapping = {'jakarta': 0, 'surabaya': 1, 'medan': 2}  # Map cuisine to numerical values
price_mapping = {'Sangat Murah': 0, 'Murah': 1, 'Mahal': 2, 'Sangat Mahal': 3}  # Map price range to numerical values

num_places = len(places)

# Create input layers
user_input = Input(shape=(1,), name='user_budget')
merchant_input = Input(shape=(1,), name='merchant_area_def')

# Embedding layers
price_embedding = Embedding(input_dim=3, output_dim=8, input_length=1, name='price_embedding')(user_input)
cuisine_embedding = Embedding(input_dim=3, output_dim=8, input_length=1, name='cuisine_embedding')(merchant_input)

# Flatten embeddings
price_flatten = Flatten()(price_embedding)
cuisine_flatten = Flatten()(cuisine_embedding)

# Concatenate flattened embeddings
concatenated_embeddings = Concatenate()([price_flatten, cuisine_flatten])

# Dense layers
dense1 = Dense(16, activation='relu')(concatenated_embeddings)
output = Dense(1, activation='sigmoid')(dense1)

# Create the model
model = Model(inputs=[user_input, merchant_input], outputs=output)

# Compile the model
model.compile(optimizer='adam', loss=MeanSquaredError())

# Prepare the input data
X_price = tf.constant([price_mapping[place['category_price']] for place in places], dtype=tf.int32)
X_cuisine = tf.constant([merchant_mapping[place['merchant_area']] for place in places], dtype=tf.int32)
y = tf.constant([1 if place['category_price'] == user_budget and place['merchant_area'] == merchant_area_def else 0 for place in places], dtype=tf.float32)

# Train the model
model.fit(x={'user_input': X_price, 'merchant_input': X_cuisine}, y=y, epochs=10)

# Generate recommendations
predictions = model.predict(x={'user_input': X_price, 'merchant_input': X_cuisine})
recommendations = [place['name'] for place, prediction in zip(places, predictions) if prediction > 0.5]

# Print the recommendations
print("Recommendations:")
for place in recommendations:
    print(places)
```
