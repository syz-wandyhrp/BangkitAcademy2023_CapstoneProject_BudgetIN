#Price Range

```python
grouped = Train.groupby('merchant_name')
grouped
```

```python
aggregated = grouped['price'].agg(['min', 'max']).reset_index()
aggregated
```

```python
aggregated["range_harga"] = aggregated['min'].astype(str) + " - " + aggregated['max'].astype(str)
aggregated
```

```python
price_ranges = {
    'Sangat Murah': (0, 15000),
    'Murah': (15000, 40000),
    'Mahal': (40000, 100000),
    'Sangat Mahal': (100000, np.inf)
}

price_ranges.items()
```

```python
def get_price_range(row):
    for category, (min_val, max_val) in price_ranges.items():
        if min_val <= row['min'] <= max_val and min_val <= row['max'] <= max_val:
            return category
    return 'Lainnya'

aggregated['kategori_harga'] = aggregated.apply(get_price_range, axis=1)
aggregated
```

```python
aggregated['merchant_area'] = Train['merchant_area']
aggregated
```

```python
kategori_mapping = {'Sangat Murah': 0, 'Murah': 1, 'Mahal': 2, 'Sangat Mahal' : 3, 'Lainnya' : 4}
area_mapping = {'Jakarta' : 0, 'Medan' : '1', 'Surabaya' : 2}

aggregated['kategori_harga'] = aggregated['kategori_harga'].map(kategori_mapping)
aggregated['merchant_area'] = aggregated['merchant_area'].map(area_mapping)
```
